```python
import cv2
import numpy as np
from ultralytics import YOLO
from collections import deque
import tensorflow as tf
import winsound
import os

# -------------------------
# CONFIG
# -------------------------
WEAPON_MODEL_PATH = "models/weapon_detection_yolov11m.pt"
VIOLENCE_MODEL_PATH = "models/cctvmodel.keras"
SEQUENCE_LENGTH = 16
ALARM_SOUND = "alarm.wav"
OUTPUT_DIR = "outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)
OUTPUT_VIDEO = os.path.join(OUTPUT_DIR, "output_tracked.mp4")

# -------------------------
# Load YOLO Model
# -------------------------
weapon_model = YOLO(WEAPON_MODEL_PATH)
weapon_model.fuse() #More Efficient

# -------------------------
# Load Keras Violence Model
# -------------------------
violence_model = tf.keras.models.load_model(
    VIOLENCE_MODEL_PATH,
    safe_mode=False,
    compile=False
)

# -------------------------
# Video Setup
# -------------------------
cap = cv2.VideoCapture(0)

fps = int(cap.get(cv2.CAP_PROP_FPS)) or 30
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(OUTPUT_VIDEO, fourcc, fps, (width, height))

frame_buffer = deque(maxlen=SEQUENCE_LENGTH)

def preprocess_frame(frame):
    img = cv2.resize(frame, (128, 128))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype("float32") / 255.0
    return img

def trigger_alarm():
    if os.path.exists(ALARM_SOUND):
        winsound.PlaySound(ALARM_SOUND, winsound.SND_ASYNC)
    else:
        winsound.Beep(1000, 500)

print("Press 'q' to quit")

# -------------------------
# Main Loop
# -------------------------
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # ----- YOLO + ByteTrack -----
    results = weapon_model.track(
        source=frame,
        persist=True,
        tracker="bytetracker.yaml", ## 3ayz add here el custom byte tracker
        conf=0.5,
        verbose=False
    )

    annotated_frame = results[0].plot()

    # Count unique tracked IDs
    weapon_count = 0
    if results[0].boxes.id is not None:
        unique_ids = set(results[0].boxes.id.cpu().numpy())
        weapon_count = len(unique_ids)

    # ----- Violence Detection -----
    frame_buffer.append(preprocess_frame(frame))
    violence_label = "Normal"

    if len(frame_buffer) == SEQUENCE_LENGTH:
        input_sequence = np.expand_dims(np.array(frame_buffer), axis=0)
        prediction = violence_model.predict(input_sequence, verbose=0)
        class_id = np.argmax(prediction)
        violence_label = "Violence" if class_id == 1 else "Normal"

    # ----- Fusion + Alarm -----
    if weapon_count > 0:
        cv2.putText(annotated_frame, "WEAPON DETECTED!", (20,40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,0,255), 3)
        trigger_alarm()

    if violence_label == "Violence":
        cv2.putText(annotated_frame, "VIOLENCE DETECTED!", (20,80),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,0,255), 3)

    cv2.putText(annotated_frame,
                f"Tracked Weapons: {weapon_count}",
                (20,120),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0,255,0),
                2)

    cv2.imshow("CCTV AI System", annotated_frame)
    out.write(annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()

print("Video saved to:", os.path.abspath(OUTPUT_VIDEO))
```
