```python
from ultralytics import YOLO
import cv2
import torch

if torch.cuda.is_available():
    print("CUDA is available!")
    print(f"CUDA Version: {torch.version.cuda}")
    print(f"Number of GPUs: {torch.cuda.device_count()}")
    print(f"Current GPU Name: {torch.cuda.get_device_name(0)}")
else:
    print("CUDA is NOT available.")
    print("Please ensure you have a compatible NVIDIA GPU and the CUDA Toolkit installed.")

# Load model
model = YOLO("models/weapon_detection_yolov11m.pt")
model.fuse()

device = 0 if torch.cuda.is_available() else "cpu"
# Open webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

print("Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (640, 480))

    # Inference
    results = model.track(frame, persist=True, conf=0.5, device=device, verbose=False,tracker="bytetrack.yaml")

    # Plot results
    annotated_frame = results[0].plot()

    # Display result
    cv2.imshow("Weapon Tracking", annotated_frame)

    # CRITICAL: This line prevents "Not Responding"
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
