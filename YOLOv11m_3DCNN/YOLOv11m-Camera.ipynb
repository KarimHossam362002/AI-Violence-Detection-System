```python
from ultralytics import YOLO
import cv2
import torch

# --------- Load Your Trained Model ----------
model = YOLO("models/weapon_detection_yolov11m.pt")
model.fuse() # More Efficient
# Optional: Use GPU if available
device = 0 if torch.cuda.is_available() else "cpu"

# --------- Open Laptop Camera ----------
cap = cv2.VideoCapture(0)  # 0 = default webcam

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

print("Press 'q' to quit.")

# --------- Tracking Loop ----------
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # ByteTrack is used here
    results = model.track(
        source=frame,
        persist=True,              # keeps IDs across frames
        tracker="custom_bytetracker.yaml",  # use ByteTrack
        conf=0.5,
        device=device,
        verbose=False
    )

    # Plot results on frame
    annotated_frame = results[0].plot()

    cv2.imshow("Weapon Tracking - YOLO + ByteTrack", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
